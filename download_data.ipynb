```python
from google.cloud import bigquery
import pandas as pd

# Initialize the BigQuery client
# You need to replace 'your-project-id' with an actual Google Cloud project ID
# For public datasets, you can use any valid project ID you have access to
client = bigquery.Client(project='your-project-id')

# Reference the dataset
dataset_ref = client.dataset("san_francisco", project="bigquery-public-data")
dataset = client.get_dataset(dataset_ref)
```

```python
# List tables in the dataset
tables = list(client.list_tables(dataset))
print(f"Tables in {dataset.dataset_id} dataset:")
for table in tables:
    print(f"- {table.table_id}")
```

```python
import os

# Example: fetch 311_service_requests table (limit for demo)
query = """
SELECT *
FROM `bigquery-public-data.san_francisco.311_service_requests`
LIMIT 10000
"""

# Execute query and convert to pandas DataFrame
df = client.query(query).to_dataframe()

# Create data directory if it doesn't exist
os.makedirs("data", exist_ok=True)

# Save locally as CSV
df.to_csv("data/311_service_requests.csv", index=False)
print(f"Downloaded {len(df)} rows to data/311_service_requests.csv")
```
